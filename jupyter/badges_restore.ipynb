```python
import datetime
import pytz

time_str = data[0]["last_edited_time"]


def check_time(time_str):
    utc_time = datetime.datetime.strptime(time_str, '%Y-%m-%dT%H:%M:%S.%fZ')

    # Установка временной зоны UTC для времени
    utc_time = utc_time.replace(tzinfo=pytz.UTC)

    # Перевод времени из UTC в московское время
    moscow_tz = pytz.timezone('Europe/Moscow')
    moscow_time = utc_time.astimezone(moscow_tz)

    # Определение промежутка времени по Москве
    start_time = moscow_tz.localize(datetime.datetime(moscow_time.year, moscow_time.month, moscow_time.day, 0, 20))
    end_time = moscow_tz.localize(datetime.datetime(moscow_time.year, moscow_time.month, moscow_time.day, 0, 45))

    # Проверка, попадает ли время в заданный промежуток
    return start_time <= moscow_time <= end_time
```

```python
# Фильтрация данных
filtered_data = []
for item in data:
    time_str = item.get('last_edited_time')
    print(time_str)
    if time_str:
        if check_time(time_str):
            filtered_data.append(item)

len(filtered_data)

# Вывод отфильтрованных данных
```
